This function is to plot the confusion matrix.


```python
def plot_confusion_matrix(y,y_predict):
    "this function plots the confusion matrix"
    from sklearn.metrics import confusion_matrix

    cm = confusion_matrix(y, y_predict)
    ax= plt.subplot()
    sns.heatmap(cm, annot=True, ax = ax); #annot=True to annotate cells
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
    ax.set_title('Confusion Matrix'); 
    ax.xaxis.set_ticklabels(['did not land', 'land']); ax.yaxis.set_ticklabels(['did not land', 'landed']) 
    plt.show() 
```

```python
## TASK  12


#Find the method performs best:


# --- SOLUCIÓN MEJORADA: Comparativa de Validación vs. Prueba ---

# Creamos un diccionario que contiene tanto la precisión de validación (best_score_) 
# como la precisión de prueba (test accuracy) para cada modelo.
results_comparison = {
    'Logistic Regression': {
        'Validation (CV)': logreg_cv.best_score_,
        'Test Data': accuracy
    },
    'SVM': {
        'Validation (CV)': svm_cv.best_score_,
        'Test Data': accuracy_svm
    },
    'Decision Tree': {
        'Validation (CV)': tree_cv.best_score_,
        'Test Data': accuracy_tree
    },
    'KNN': {
        'Validation (CV)': knn_cv.best_score_,
        'Test Data': accuracy_knn
    }
}

print(f"{'Model':<20} | {'Validation Acc':<15} | {'Test Acc':<10}")
print("-" * 50)

# Buscamos el mejor modelo basándonos en la precisión de TEST (que es lo que importa para generalizar)
best_model_test = None
max_test_acc = -1

for model, metrics in results_comparison.items():
    val_acc = metrics['Validation (CV)']
    test_acc = metrics['Test Data']
    
    print(f"{model:<20} | {val_acc:<15.4f} | {test_acc:<10.4f}")
    
    if test_acc > max_test_acc:
        max_test_acc = test_acc
        best_model_test = model

print("-" * 50)
print(f"RESULTADO FINAL:")
print(f"El modelo que mejor generaliza (mayor precisión en TEST) es: {best_model_test}")
# ---------------------------------------------------------
```
